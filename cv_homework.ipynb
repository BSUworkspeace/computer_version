```python
NpKernel = np.zeros((3,5))
for i in range(5):
    NpKernel[0, i] = -1 
    NpKernel[1, i] = 0
    NpKernel[2, i] = 1
NpKernel[1,1]=1
```

```python

def filter2d(image, kernel, bound):
    m, n = kernel.shape
    y,x,z= image.shape
    y_out = y - m + 1
    x_out  = x - n + 1
#     print(m,n,y,x,z)
    new_image = np.zeros((y_out, x_out))
    for i in range(y_out):
        for j in range(x_out):
            new_image[i][j] = np.sum(image[i:i+m, j:j+n]*kernel[:,:,None])
    return new_image
```
